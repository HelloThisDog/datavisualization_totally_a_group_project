```python
df_clean = duckdb.sql("""--sql
SELECT
    * EXCLUDE(Priority, Satisfaction),
    CASE
        WHEN Priority = '0 - Unassigned' THEN 'Unassigned'
        WHEN Priority = '1 - Low' THEN 'Low'
        WHEN Priority = '2 - Medium' THEN 'Medium'
        WHEN Priority = '3 - High' THEN 'High'
    END AS Priority
    CASE
        WHEN Satisfaction = '0 - Unknown' THEN 'Unknown'
        WHEN satisfaction = '1 - Unsatisfied' THEN 'Unsatisfied'
        WHEN Satisfaction = '2 - Satisfied' THEN 'Satisfied'
        WHEN satisfaction = '3 - Highly satisfied' THEN 'Highly satisfied'
FROM df
""").df()
```

```python
df_clean.to_csv("IT-Help-clean.csv")
```
